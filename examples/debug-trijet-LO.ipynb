```python
import numpy as np
from uncertainties import unumpy as unp
def fitr(x):
    return 1./(unp.exp(-10*(x-0.5))+1)
for p in [dddf.xs("pwg-btlgrid").xs(1)["plot"][0]]:
    p.fit(fitr,size=(5,5),fmt="-",logx=False,fselector=lambda x,y : np.logical_and(x>0,x>0) ,extrapolate=False)
```
